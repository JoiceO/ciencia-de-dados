##2. Tokenizando o *corpus*


```
import nltk
from nltk import tokenize

nltk.download('punkt') # necessário para baixar o tokenizador
```

Armazenando as sentenças.

```
poema = poema.lower().split("\n")
poema
```

```
texto_tok = []
for verso in poema:
  tokens = tokenize.word_tokenize(verso, language="portuguese")
  texto_tok.append(tokens)

texto_tok
```

##3. Inserindo os marcadores de início e fim de sentença

Usando um método do NLTK que se chama `pad_both_ends`

Bigramas

```
from nltk.lm.preprocessing import pad_both_ends

ngramas = 2

texto_tok_pad = []

for verso in texto_tok:
  padded = list(pad_both_ends(verso,n=ngramas))
  texto_tok_pad.append(padded)

texto_tok_pad
```

##4. Calculando os n-gramas

Usando um método do NLTK que se chama `nltk.ngrams`

```
ngramas = 2

bigramas_poema = []
for verso in texto_tok_pad:
  bigramas = list(nltk.ngrams(verso, n=ngramas))
  bigramas_poema.append(bigramas)

bigramas_poema
```

Mas para gerar o modelo de linguagem é preciso que os n-gramas de tamanho menor também estejam definidos. Para isso, é mais simples usar o método `nltk.everygrams` que já cuida desse "detalhe"



```
from nltk.util import everygrams

tam_ngrama = 3

ngramas_poema = []
for verso in texto_tok_pad:
  ngramas = list(everygrams(verso, max_len=tam_ngrama))
  ngramas_poema.append(ngramas)

ngramas_poema
```

##5. Colocando todos os *tokens* do *corpus* em uma única lista

Para isso vamos usar o método `nltk.lm.preprocessing.flatten` que junta os elementos de todas as listas em uma única.

```
from nltk.lm.preprocessing import flatten

tokens = list(flatten(texto_tok_pad))
tokens
```

##6. Definindo o vocabulário

Para isso vamos usar o `nltk.lm.Vocabulary` e o parâmetro indica a frequência mínima, o  `unk_cutoff`

```
from nltk.lm import Vocabulary

vocab = Vocabulary(tokens, unk_cutoff=1)
vocab.counts
```

## Simplificando o processo (passos 3 a 6 com um só comando)

```
from nltk.lm.preprocessing import padded_everygram_pipeline

ngrams = 3
ngramas_pad, vocab = padded_everygram_pipeline(ngrams, texto_tok)
```

##7. Treinando um modelo de linguagem

Usamos o método `nltk.lm.MLE`

```
from nltk.lm import MLE

ngrams = 3
lm = MLE(ngrams)
lm.fit(ngramas_pad,vocab)
```

Vamos gerar trechos de texto usando o método `lm.generate` que recebe como parâmetro o tamanho da sequência que a gente deseja gerar e o contexto, ou seja, com base em algum *token*.

```
#lm.generate(5,text_seed=["você"])
lm.generate(5,text_seed=["você"], random_seed = 3)
```

Verificar a probabilidade de uma palavra.

```
lm.score(",")
```

```
lm.score("agora", context=["e"])
```

#8. Avaliação



##Cálculo de perplexidade

```
teste = """E agora, José?
A festa apagou,
a luz acabou
"""

# pré-processamento
teste = teste.lower().split('\n')
teste_tok = []
for verso in teste:
  tokens = nltk.word_tokenize(verso, language='portuguese')
  teste_tok.append(tokens)

ngramas = 1
teste_ngramas, _ = padded_everygram_pipeline(ngramas, teste_tok)
teste_ngramas = flatten([list(w) for w in teste_ngramas])
print("Perplexidade do Unigrama: ", lm.perplexity(teste_ngramas))

ngramas = 2
teste_ngramas, _ = padded_everygram_pipeline(ngramas, teste_tok)
teste_ngramas = flatten([list(w) for w in teste_ngramas])
print("Perplexidade do Bigrama: ", lm.perplexity(teste_ngramas))
```

## Suavização



* Add-1 Smoothing

```
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import Laplace

ngramas = 2
ngramas_pad, vocab = padded_everygram_pipeline(ngramas, texto_tok)
lm = Laplace(ngramas)
lm.fit(ngramas_pad, vocab)
```

```
round(lm.score("agora", context=["e"]), 2)
```

* Add-k Smoothing

```
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import Lidstone

ngramas = 2
k=0.1
ngramas_pad, vocab = padded_everygram_pipeline(ngramas, texto_tok)
lm = Lidstone(order=ngramas, gamma=k)
lm.fit(ngramas_pad, vocab)
```

```
round(lm.score("agora", context=["e"]), 2)
```

#9. Representações Vetoriais

## Representação *One-Hot*
Palavras e documentos são representados por vetores de dimensão do tamanho do vocabulário. Os vetores assumem valores binários (0 ou 1).

Vamos usar a biblioteca `OneHotEncoder` do Scikit Learn.

```
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')

X = [["o"], ["menino"], ["foi"], ["para"], ["a"], ["escola"], ["de"], ["ônibus"]]

enc.fit(X)

vocab = list(enc.categories_[0])
vetores = enc.transform(X).toarray()

print("Vocabulário\n", vocab, "\n")
print("Matriz\n", vetores, "\n")

print("Atributos\n", enc.get_feature_names_out()) # mesma ordem de vocab
```

## Matriz Frequência Termo-Documento (*Bag of Words*)
Dado um vocabulário e um conjunto de documentos, as representações das palavras e dos documentos podem ser calculadas a partir da contagem de cada palavra em cada documento.

Para isso vamos usar o método `CountVectorizer` do Scikit Learn.

```
import nltk
nltk.download('punkt') # necessário para baixar o tokenizador

from sklearn.feature_extraction.text import CountVectorizer
```

Carregando o *corpus*.

```
#poema = ["E agora, José?", "A festa acabou,", "a luz apagou,", "o povo sumiu,", "a noite esfriou,", "e agora, José?", "e agora, você?", "você que é sem nome,",
#          "que zomba dos outros,", "você que faz versos,", "que ama, protesta?", "e agora, José?"]
poema = ["A festa acabou e apagou quando a luz apagou e acabou,", "A luz acabou", "E agora, José?", "A festa acabou,", "a luz apagou,", "o povo sumiu,", "a noite esfriou,", "e agora, José?", "e agora, você?", "você que é sem nome,",
          "que zomba dos outros,", "você que faz versos,", "que ama, protesta?", "e agora, José?"]
print(len(poema))
```

Gerando a matriz de frequência termo-documento.

```
def tokenize(texto):
  return nltk.word_tokenize(texto, language="portuguese")

vectorizer = CountVectorizer(tokenizer=tokenize,lowercase=True)

vetores = vectorizer.fit_transform(poema)
vocab = vectorizer.get_feature_names_out()

print("Vocabulário\n", vocab,"\n")
print("Matriz\n", vetores.toarray(), "\n")
```

## Matriz de Frequência Termo-Termo

```
import numpy as np

corpus_tok = []
for verso in poema:
  tokenizado = str(tokenize(verso.lower()))
  corpus_tok.append(tokenizado)

vocab = ['a', 'acabou', 'agora', 'ama', 'apagou', 'dos', 'e', 'esfriou', 'faz', 'festa', 'josé', 'luz',
  'noite', 'nome', 'o', 'outros', 'povo', 'protesta', 'que', 'sem', 'sumiu', 'versos', 'você', 'zomba', 'é']

vetores = np.zeros((len(vocab), len(vocab)))

for verso in corpus_tok:
  for i, w1 in enumerate(vocab):
    for j, w2 in enumerate(vocab):
      if i != j:
        if w1 in verso and w2 in verso:
          vetores[i, j] += 1

print("Vocabulário\n", vocab,"\n")
print("Matriz\n", vetores)
```

## Remoção de *stopwords*
Reduzindo a esparsidade da representação removendo *stopwords*.

```
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('portuguese')

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words=stopwords)

vetores = vectorizer.fit_transform(poema)
vocab = vectorizer.get_feature_names_out()

print("Vocabulário\n", vocab , "\n")
print("Matriz\n", vetores.toarray())
```

## TF-IDF

```
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline

vectorizer = Pipeline([('count', CountVectorizer()),
                 ('tfid', TfidfTransformer())])

vetores = vectorizer.fit_transform(poema)
vocab = vectorizer['count'].get_feature_names_out()

print("Vocabulário\n", vocab, "\n")
print('Matriz\n', np.round(vetores.toarray(), 2))
```

Acessando o oitavo ("e agora, José?") e o nono ("e agora, você?") versos e calculando a similaridade entre eles.

```
from sklearn.metrics.pairwise import cosine_similarity

id_verso_a = 7
id_verso_b = 8

verso_a = vetores[id_verso_a, :]
print("Verso A = ", poema[id_verso_a])
verso_b = vetores[id_verso_b, :]
print("Verso B = ", poema[id_verso_b])

cosine_similarity(verso_a,verso_b)[0][0]
```
